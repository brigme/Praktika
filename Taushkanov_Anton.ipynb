### Оценка популярности книги по количеству отзывов

```python
df['new_column'] = np.where(df.reviews < 10000, 'Not popular', 'Popular')
df
```

### Расчет скидки, если цена больше 20$

```python
df['discount'] = np.where(df.price > 20, df.price * 0.5, df.price)
df
```

```python
df.drop(labels = 'discount', axis = 1, inplace = True)
```

### Есть ли необходимость сделать скидку из-за непопулярности книги

```python
df['new_column'] = np.where((df.reviews < 5000) & (df.price >= 15), 'Need discount','Good price' )
df
```

### Популярность книги

```python
df['new_column'] = np.where((df.reviews > 20000) & (df.user_rating >= 4.8), 'Good and popular','Not popular' )
df
```

# ***Графический анализ данных***

## **1. Диаграммы, библиотека Matplotlib**

### Средняя цена по годам

```python
avg_price = pd.pivot_table(df, index = 'year', values = 'price', aggfunc = 'mean')
avg_price
```

```python
avg_price.plot(kind = 'line', legend = True, title = 'Средняя цена по годам', marker = 'o')
```

### 20 авторов, которые чаще всех входили в список лучших

```python
author_count = pd.pivot_table(df, index = 'author', values = 'name',
                              aggfunc = 'count').sort_values(by = 'name',ascending = False)[:20]
```

```python
author_count.rename(columns = {'name': 'count'}, inplace = True)
```

```python
author_count.plot(kind = 'bar', title = '20 авторов, которые чаще всех входили в список лучших')
plt.xlabel('Автор')
plt.ylabel('Сколько раз входил в список')
```

### Соотношение художественной и нехудожественной литературы

```python
genre_sum = pd.pivot_table(df, index = 'genre', values = 'name', aggfunc = 'count').rename(columns = {'name': 'sum_genre'})
```

```python
genre_sum.plot(autopct='%1.1f%%', kind = 'pie', legend=True, figsize=(8, 6), subplots=True,
               title = 'Соотношение художественной и нехудожественной литературы')
```

## **2. Диаграммы, библиотека Seaborn**

### Сравнение количества жанров в списке по годам

```python
genre_by_year = df.groupby(by = ['year', 'genre']).count()
genre_by_year
```

```python
genre_year = sns.barplot(data = genre_by_year, x = 'year', y = 'name', hue = 'genre')
genre_year.grid(True)
plt.title('Сравнение количества жанров в списке по годам')
genre_year.set_xlabel('Год')
genre_year.set_ylabel('Сколько раз вошли в список')
```

### Самый популярный ценовой сегмент

```python
scat = sns.scatterplot(data = df, x = 'price', y = 'reviews')
plt.title('Самый популярный ценовой сегмент')
scat.set_xlabel('Цена')
scat.set_ylabel('Кол-во отзывов')
```
